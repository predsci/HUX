# Apply PDE-FIND algorithm to find the underlying PDE using sparse regression. 

The solar wind motion can be described as the fluid momentum equation in corotating frame of reference:
	
$$
-\Omega_{rot} \frac{\partial v}{\partial \phi} + (v \cdot \nabla)v = \frac{1}{\rho} \nabla P - \frac{G M_{s}}{r^2}e_{r}
$$

By neglecting magnetic field, pressure gradient, and gravity, the fluid momentum equation reduces to the inviscid Burgers' equation.

$$
\frac{\partial v_{r}}{\partial \phi} = \frac{1}{\Omega_{rot}} v_{r} \frac{\partial v_{r}}{\partial r}
$$


$${\Omega_{rot}}_{\theta = \pi/2} = \frac{2\pi}{25.38} \frac{1}{days} = \frac{2\pi}{25.38\cdot 86400} \frac{1}{sec}$$

The goal is to analyze the significance of each term in the fluid momentum equation by leveraging a data driven approach. 

The HUX model neglects thermal pressure and proton mass density from the fluid momentum equation. In order, to understand the weight of adding thermal pressure and proton mass density to forecast solar wind streams, we applied PDE-FIND a parsimonious algorithm developed by Rudy et al [2]. Given a large library of candidate terms consisting partials and nonlinearities in spatial domain, PDE-FIND finds the optimal subset of active library terms from a time sequential data-set.


$$
\frac{\partial v_{r}}{\partial \phi} = \xi \Theta(\frac{\partial v_{r}}{\partial r}, v_{r}\frac{\partial v_{r}}{\partial r}, \frac{\partial^2 v_{r}}{\partial r^2}, v_{r}^2\frac{\partial^2 v_{r}}{\partial r^2}, v_{r}^3,\frac{1}{\rho} \nabla P, ...)
$$
    
$\Theta$ represents the library of potential terms, $\xi$ is a sparse vector containing the optimal coefficients. In order to build $\Theta$ library, there are several ways to numerically differentiate. A common technique is using central finite differences. Given a noisy dataset the numerical derivatives will be dominated by the effect of noise, and instead a better method is polynomial interpolation. In our case, the solar wind dataset has no noise (since we are using the MHD solutions as 'ground truth'), therefore we use finite differences method. 


# References
[1] [Riley, P. and Lionello, Roberto. Mapping solar wind streams from the Sun to 1 AU: A comparison of techniques. Solar Physics, 270(2), 575–592, 2011.](https://www.researchgate.net/publication/226565167_Mapping_Solar_Wind_Streams_from_the_Sun_to_1_AU_A_Comparison_of_Techniques)

[2] [Samuel H. Rudy, Steven L. Brunton, Joshua L. Proctor, and J. Nathan Kutz. Data-driven discovery of partial differential equations. Science Advances, 3(4):e1602614, 2017.](https://arxiv.org/abs/1609.06401)


## Sparse Regression

This linear system is solved using sparse regression. Sparse regression penalizes large models and therefore performs variable selection. Some of the most common sparse regression algorithms used in machine learning are LASSO and ridge, which are regularization techniques.

- LASSO: L1 NORM. 


$${\displaystyle \min _{\beta }\left\{\sum _{i=1}^{N}(y_{i}-x_{i}^{T}\beta )^{2}\right\}{\text{ subject to }}\sum _{j=1}^{p}||\beta _{j}||_{1}\leq t.}$$

- Ridge: L2 NORM. 

$${\displaystyle \min _{\beta }\left\{\sum _{i=1}^{N}(y_{i}-x_{i}^{T}\beta )^{2}\right\}{\text{ subject to }}\sum _{j=1}^{p}||\beta _{j}||_{2}\leq t.}$$

- STLS: Sequentially thresholded least squares. 

$${\displaystyle \min _{\beta }\left\{\sum _{i=1}^{N}(y_{i}-x_{i}^{T}\beta )^{2}\right\}{\text{ subject to }}\sum _{j=1}^{p}||\beta _{j}||_{2}^{2}\leq t.}$$

The objective:

$$\zeta^{*} = argmin_{\zeta} || \Theta \zeta - U_{t}||_{2}^{2} + \lambda ||\zeta||_{2}^{2}= (\Theta^{T} \Theta + \lambda I)^{-1} \Theta^{T} U_{t}$$

Supplemental notes observations: 
    
    * LASSO tends to have difficulty finding a sparse basis when the data matrix Θ has high correlations between columns

# Thermal Pressure dataset.

### The fluid momentum equation in corotating frame of reference

$$-\Omega_{rot} \frac{\partial v}{\partial \Phi} + (v \cdot \nabla)v = \frac{1}{\rho} \nabla P - \frac{G M_{s}}{r^2}e_{r}$$


By neglecting the gravity term, we get: 

$$-\Omega_{rot} \frac{\partial v}{\partial \Phi} + (v \cdot \nabla)v = \frac{1}{\rho} \nabla P$$

$$-\Omega_{rot} \frac{\partial v}{\partial \Phi} + v \frac{\partial v}{\partial r} = \frac{1}{\rho} \frac{ \partial P}{\partial r}$$

$$\frac{\partial v}{\partial \Phi} =\frac{1}{\Omega_{rot}}  v \frac{\partial v}{\partial r} - \frac{1}{\Omega_{rot}}\frac{1}{\rho} \frac{ \partial P}{\partial r}$$


units in MKS (for normalization): 

- $\Omega_{rot} - \frac{radians}{sec} - \frac{radians}{days}$


- $ v - \frac{km}{sec} - \frac{Solar Radii}{days}$

- $ \Phi - radians$

- $ r - km - Solar Radii$

- $ \rho (density) - \frac{kg}{m^3} - \frac{kg}{Solar Radii^3}$

- $ P (pressure) - 1 Pa - \frac{kg}{ms} - \frac{kg}{Solar Radii \cdot days}$

# Step 2: Construct $\Theta (U)$ and compute partials. 


$$
\frac{\partial v_{r}}{\partial \phi} = \xi \Theta(\frac{\partial v_{r}}{\partial r}, v_{r}\frac{\partial v_{r}}{\partial r}, \frac{\partial^2 v_{r}}{\partial r^2}, v_{r}^2\frac{\partial^2 v_{r}}{\partial r^2}, v_{r}^3,\frac{1}{\rho} \nabla P, ...)
$$
    

```python
dr_avg = np.mean(dr)
dp_avg = np.mean(dp)

print("Average dr", dr_avg)
print("Average dp", dp_avg)
```

```python
u_t = np.gradient(slice_at_equator, dp_avg, axis=0).flatten()
u_x = np.gradient(slice_at_equator, dr_avg, axis=1).flatten()
```

```python
# build uux. 
ux = np.gradient(slice_at_equator, dr_avg, axis=1)
uxx = np.gradient(ux, dr_avg, axis=1)
u = slice_at_equator
uux = np.zeros((u.shape[0], u.shape[1]))
u2ux = np.zeros((u.shape[0], u.shape[1]))
u2 = np.zeros((u.shape[0], u.shape[1]))
uuxx = np.zeros((u.shape[0], u.shape[1]))
u2uxx = np.zeros((u.shape[0], u.shape[1]))
for ii in range(u.shape[0]):
    for jj in range(u.shape[1]):
        uux[ii, jj] = u[ii, jj] * ux[ii, jj]
        u2[ii, jj] = u[ii, jj]**2
        uuxx[ii, jj] = u[ii, jj] * uux[ii, jj]
        u2uxx[ii, jj] = u2[ii, jj] * uux[ii, jj]
        u2ux[ii, jj] =  u2[ii, jj] * ux[ii, jj]
```

```python
lhs = u_t # partial with respect to phi. 
rhs = np.zeros((len(lhs), 10)) # up to order 2, 9 elements. 
# Meaning: 1, u_x, uu_x, u^2u_x, u_xx, uu_xx, u^2u_xx, u^2, u
rhs[:, 0] = np.zeros(len(lhs))
rhs[:, 1] = ux.flatten()
rhs[:, 2] = uux.flatten()
rhs[:, 3] = u2ux.flatten()
rhs[:, 4] = uxx.flatten()
rhs[:, 5] = uuxx.flatten()
rhs[:, 6] = u2uxx.flatten()
rhs[:, 7] = u2.flatten()
rhs[:, 8] = u.flatten()
rhs[:, 9] = extra_element.flatten()
```

# Step 3: find the optimal combination using ridge sparse regression. 
# $\xi^{*} = argmin_\xi || \frac{\partial v_{r}}{\partial \phi} - \xi \Theta||_{2}^2 + \lambda ||\xi||_{2}^2$

```python
sol = scipy.sparse.linalg.lsqr(A=rhs, b=lhs, damp=20, atol=1e-08, btol=1e-08, conlim=100000000.0, iter_lim=None, show=True, calc_var=False, x0=None)
```

```python
fig, ax = plt.subplots()

_ = ax.scatter(np.arange(len(sol[0])), sol[0])
_ = ax.set_ylabel("coefficient", fontsize=17)
_ = ax.set_xlabel("term index", fontsize=17)
_ = ax.set_title("$\\xi$ result", fontsize=20)
```

```python
sol[0]
```

## Notice that the second and last term are the most dominant. 
- The second element = $\partial v_{r}/\partial \phi$
- The last term = $∇P/p$

```python
# Compute the exact c2 and c9 coefficients. 
rhs[:, 0] = np.zeros(len(lhs))
rhs[:, 1] = np.zeros(len(lhs))
rhs[:, 2] = uux.flatten()
rhs[:, 3] = np.zeros(len(lhs))
rhs[:, 4] = np.zeros(len(lhs))
rhs[:, 5] = np.zeros(len(lhs))
rhs[:, 6] = np.zeros(len(lhs))
rhs[:, 7] = np.zeros(len(lhs))
rhs[:, 8] = np.zeros(len(lhs))
rhs[:, 9] = extra_element.flatten()
```

```python
sol = scipy.sparse.linalg.lsqr(A=rhs, b=lhs, damp=20, atol=1e-08, btol=1e-08, conlim=100000000.0, iter_lim=None, show=True, calc_var=False, x0=None)
```

```python
print("c1 = " , sol[0][2])
```

```python
print("c2 = ", sol[0][-1])
```

```python
rhs_res_1 =  sol[0][2]*rhs[:, 2] + sol[0][-1]*rhs[:, -1]
```

```python
mse_res1 = (np.square(lhs - rhs_res_1)).mean()
```

```python
# plot the solution and observe if the rhs equals the lhs.
_ = plt.plot(rhs_res_1, label="$c_{1}\partial v_{r}/\partial \phi +c_{2} ∇P/p$")
_ = plt.plot(u_t, label="$\partial v_{r}/ \partial \phi$", color='r')
_ = plt.legend(fontsize=20)
_ = plt.ylabel("RS/day", fontsize=20)
_ = plt.xlabel("index-flatten", fontsize=20)
_ = plt.title("PDE fit, MSE =" + str(round(mse_res1, 3)), fontsize=20)
```

# Results summary 

Ridge results validate that the terms $v_{r}\frac{\partial v_{r}}{\partial r}$ and $\frac{1}{\rho} \nabla P$ are the most dominant terms in the library. The resulting equation for CR 2068 is  $\frac{\partial v_{r}}{\partial \phi} = 3.59v_{r}\frac{\partial v_{r}}{\partial r} + 3.14\frac{1}{\rho} \nabla P$, with pearson coefficient of 0.977 and mean squared error of 136.012. Whereas, when excluding $\frac{1}{\rho} \nabla P$ from the library $\Theta$, the resulting underlying equation is $\frac{\partial v_{r}}{\partial \phi} = 3.398v_{r}\frac{\partial v_{r}}{\partial r}$, with pearson coefficient of 0.956 and mean squared error of 296.149. Therefore, by adding $\frac{1}{\rho} \nabla P$ to model solar wind proton velocity, we reduce the model's relative error by 8.65 percent.
